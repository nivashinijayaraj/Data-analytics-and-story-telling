```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder 
```

```python
le = LabelEncoder() 
ds = pd.DataFrame(data = {'col1': ['foo','bar','foo','bar'], 
'col2': ['x', 'y', 'x', 'z'], 
'col3': [1, 2, 3, 4]}) 
```

```python
ds.apply(le.fit_transform)
```

```python
ds=pd.DataFrame({'A':['a','b','a'],'B':['b','a','c'],'C':[1,2,3]}) 
ds 
```

```python
pd.get_dummies(ds,prefix=['col1','col2']) 
```

```python
from sklearn.preprocessing import MinMaxScaler 
mm_scaler=MinMaxScaler(feature_range=(0,1)) 
ds2=pd.DataFrame({"col1":[5,-41,-67],"col2":[23,-53,-36],"col3":[-25,10,17]}) 
mm_scaler.fit_transform(ds2)
```

```python
from sklearn.preprocessing import Binarizer 
dsb=pd.DataFrame({"col1":[110,200],"col2":[120,800],"col3":[310,400]}) 
bin=Binarizer(threshold=300) 
bin.fit_transform(dsb) 
```

```python
import numpy as np 
from sklearn.impute import SimpleImputer 
import pandas as pd 
imp_mean=SimpleImputer(missing_values=np.nan,strategy='mean') 
dr=pd.DataFrame({"col1":[7,2,3], 
 "col2":[4,np.nan,6], 
 "col3":[np.nan,np.nan,3], 
 "col4":[10,np.nan,9] }) 
print(dr) 
imp_mean.fit_transform(dr) 
```
